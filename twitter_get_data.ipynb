```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

import time
import csv
import pandas as pd
```

Google üzerinde basit arama örneği

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.google.com/")
arama=driver.find_element("css selector",".gLFyf.gsfi")
arama.send_keys("selenium nedir")
time.sleep(1)
arama.send_keys(Keys.ENTER)
```

Twitter üzerinden arama sonuçlarını alma

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))
#driver.get("https://twitter.com/search?q=yapay%20zeka")
  
driver.get("https://www.google.com.tr/")
yazı_girişi = driver.find_element("css selector",".gLFyf.gsfi")
yazı_girişi.send_keys("twitter deep learning türkiye")
time.sleep(2)
yazı_girişi.send_keys(Keys.ENTER)

tıkla = driver.find_element("css selector",".yuRUbf")
tıkla.click()
    
kaynak = driver.page_source
bs = BeautifulSoup(kaynak, "html.parser")
tweets = bs.find_all("div",attrs={"data-testid":"tweet"})
for twt in tweets:
    try:
         text = twt.find("div", attrs={"class":"css-901oao r-hkyrab r-1qd0xha r-a023e6 r-16dba41 r-ad9z0x r-bcqeeo r-bnwqim r-qvutc0"}).text
         yanit = twt.find("div", attrs={"data-testid":"reply"}).text
         retweet = twt.find("div", attrs={"data-testid":"retweet"}).text
         begeni = twt.find("div", attrs={"data-testid":"like"}).text
         print(text)
         print("yanıt sayısı=", yanit)
         print("retweet sayısı=", retweet)
         print("beğeni sayısı=", begeni)
                #writer.writerow([yazı,beğeni_sayısı,yorum_sayısı,retweet_sayısı])
    except:
         print("")
```
